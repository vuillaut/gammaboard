```python
import numpy as np

def create_plot_on_click(experiments_dict, fig_resolution): #, visible_experiments, ax_exp, ax_auc, site='south'):

    def plot_on_click(sender):
        """
        Function to be called when a `ipywidgets.Button` is clicked

        Args
            sender: the object received by `ipywidgets.Button().on_click()`
        """
        exp_name = sender.description
        if exp_name not in experiments_dict:
            pass

        exp = experiments_dict[exp_name]

#         if sender.button_style == 'warning':
#             sender.button_style = 'success'
#             visible = True
#             if not exp.get_loaded():
#                 exp.load_data()
#             if exp_name not in tabs.keys():
#                 tabs[exp_name] = Output()
#             with tabs[exp_name]:
#                 try:
#                     print_dict(exp.info)
#                 except:
#                     print("Sorry, I have no info on the experiment {}".format(exp_name))
#             visible_experiments.add(exp)
#         else:
#             sender.button_style = 'warning'
#             visible = False
#             tabs[exp_name].close()
#             tabs.pop(exp_name)
#             visible_experiments.remove(exp)

        if not exp.get_plotted() and visible and exp.data is not None:
            plot_exp_on_fig(exp, fig_resolution, site)

        exp.visibility_all_plot(visible)
        update_legend(visible_experiments, ax_exp)
        update_auc_legend(visible_experiments, ax_auc)

    return plot_on_click


def slider_data(column_name, log=False):
    col = exp.data[column_name]
    mi = np.round(col.min(), decimals=2)
    ma = np.round(col.max(), decimals=2)
    if log:
        mi = np.round(np.log10(col.min()), decimals=2)
        ma = np.round(np.log10(col.max()), decimals=2)
    step = np.round((ma - mi)/10., decimals=2)
    options = np.arange(mi, ma+step, step)
    return SelectionRangeSlider(options=options, value=(options.min(), options.max()))


def make_experiments_carousel(experiments_dic, fig_resolution): #, visible_experiments, ax_legend, ax_auc, site):
    """
    Make an ipywidget carousel holding a series of `ipywidget.Button` corresponding to
    the list of experiments in experiments_dic
    Args
        experiments_dic (dict): dictionary of experiment class
        experiment_info_box (Tab): the tab container
        tabs (dict): dictionary of active tabs
        fig_resolution
        visible_experiments
        ax_legend
        ax_auc

    Returns
        `ipywidgets.VBox()`
    """
    from ipywidgets import Layout, Button, VBox, Dropdown, SelectionRangeSlider
    
#     return Dropdown(options=exp.data.columns)

    item_layout = Layout(width='auto')

    dd = Dropdown(options=exp.data.columns)
    slider = slider_data('disp_norm')
    dd.observe(slider_data, names='value')
    items = [dd,
             slider,
            ]
    
# #     items = [Button(layout=item_layout, description=exp_name, button_style='warning')
# #              for exp_name in np.sort(list(experiments_dic))[::-1]]
#     items = [Button(layout=item_layout, description=c, button_style='warning') for c in exp.data.columns]

#     for b in items:
#         b.on_click(create_plot_on_click(experiments_dic, fig_resolution)) #, visible_experiments, ax_legend, ax_auc, site))

    box_layout = Layout(
                        # border='3px solid black',
#                         width='300px',
#                         height='600px',
                        flex_flow='row',
                        display='flex')

    return VBox(children=items, layout=box_layout)
```

```python
%matplotlib inline
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np


vardict = exp.data.columns
select_variable = widgets.Dropdown(
    options=vardict,
    value=vardict[0],
    description='Select variable:',
    disabled=False,
    button_style=''
)

logbox = widgets.Checkbox(description='log')

slider = SelectionRangeSlider(options=[-np.inf,np.inf], value=(-np.inf, np.inf))


def get_and_plot(b):
#     clear_output
    column_name = select_variable.value
#     slider.description=column_name
    if logbox.value == False:
        mi = np.round(exp.data[column_name].min(), decimals=2)
        ma = np.round(exp.data[column_name].max(), decimals=2)
    else:
        mi = np.round(np.log10(exp.data[column_name].min()), decimals=2)
        ma = np.round(np.log10(exp.data[column_name].max()), decimals=2)
    opt = np.round(np.linspace(mi, ma), decimals=2)
    slider.options=opt
    slider.value=(mi, ma)
#     print(select_variable.value)

display(select_variable, logbox)
select_variable.observe(get_and_plot, names='value')
logbox.observe(get_and_plot, names='value')
# slider.observe(f, names='value')


def f(b):
    fig = plt.figure()
    exp.mask = (exp.data[select_variable.value] > slider.value[0]) & (exp.data[select_variable.value] < slider.value[1])
    E, R = ctaplot.ana.angular_resolution_per_energy(exp.data[exp.mask].reco_altitude,
                                         exp.data[exp.mask].reco_azimuth,
                                         exp.data[exp.mask].mc_altitude,
                                         exp.data[exp.mask].mc_azimuth,
                                         exp.data[exp.mask].mc_energy,)

    ctaplot.plot_resolution(np.log10(E), np.rad2deg(R))
#     ax = ctaplot.plot_angular_res_per_energy(exp.data[exp.mask].reco_altitude,
#                                          exp.data[exp.mask].reco_azimuth,
#                                          exp.data[exp.mask].mc_altitude,
#                                          exp.data[exp.mask].mc_azimuth,
#                                          exp.data[exp.mask].mc_energy,
#                                         ax=axes
#                                        )
    #     plt.plot(np.log10(E[:-1]), R)
    plt.ylim(0, 1)
    plt.show(fig)


# interactive_plot = interactive(f)
# output = interactive_plot.children[-1]
# output.layout.height = '350px'
# interactive_plot
interactive_plot = interactive(f, b=slider)
interactive_plot
```
